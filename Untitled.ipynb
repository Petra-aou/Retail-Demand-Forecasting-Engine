```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. 读取 Kaggle 真实数据 (请确保路径和你的本地路径一致)
# 真实数据里包含：id, date, store_nbr, family, sales, onpromotion
train = pd.read_csv('train.csv', parse_dates=['date'])

# 2. 真实赛题的数据颗粒度非常细（包含各种商品类别 family 和不同门店 store_nbr）
# 为了做经典的时间序列检验，我们先按“天”对全超市的总销量进行聚合（Aggregation）
daily_sales = train.groupby('date')['sales'].sum().to_frame()

# 3. 绘制真实的零售总销量时序图
plt.figure(figsize=(12, 5))
plt.plot(daily_sales.index, daily_sales['sales'], color='blue', linewidth=1)
plt.title("Total Daily Store Sales (Kaggle Real Data)")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.grid(True)
plt.show()
```

```python
# 1. 核心数学动作：今天减昨天
daily_sales['sales_diff'] = daily_sales['sales'].diff()

# 2. 【避坑标准写法】：建立一个干净的临时表，把第一天的 NaN 整行删掉，这样 x 和 y 永远维度一致
cleaned_df = daily_sales.dropna()

# 3. 绘图（此时 x 和 y 都来自 cleaned_df，绝对不会报错）
plt.figure(figsize=(12, 5))
plt.plot(cleaned_df.index, cleaned_df['sales_diff'], color='darkgreen', linewidth=1)

plt.title("Differenced Daily Store Sales (Stationary Stream)")
plt.xlabel("Date")
plt.ylabel("Sales Delta (Change)")
plt.grid(True)
plt.show()
```

```python
!pip install statsmodels
from statsmodels.tsa.stattools import adfuller

# 对你刚才亲手洗干净、切掉空值的绿色 sales_diff 序列进行 ADF 检验
adf_result = adfuller(cleaned_df['sales_diff'])

print(f"ADF Statistic (统计量): {adf_result[0]:.4f}")
print(f"p-value (P值): {adf_result[1]:.4f}")
```

```python
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# 创建一个并排的画布，左边放 ACF，右边放 PACF
fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# 1. 画出 sales_diff 的自相关图 (ACF) —— 用来决定 q 参数
plot_acf(cleaned_df['sales_diff'], lags=30, ax=axes[0])
axes[0].set_title("Autocorrelation (ACF) - Guess q")

# 2. 画出 sales_diff 的偏自相关图 (PACF) —— 用来决定 p 参数
plot_pacf(cleaned_df['sales_diff'], lags=30, ax=axes[1])
axes[1].set_title("Partial Autocorrelation (PACF) - Guess p")

plt.show()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
import pandas as pd
import matplotlib.pyplot as plt

# 1. 填入 pmdarima 帮你想出来的终极王牌参数！
final_model = SARIMAX(cleaned_df['sales'], 
                      order=(3, 1, 2), 
                      seasonal_order=(2, 0, 2, 7),
                      trend='c') # 'c' 代表 constant，也就是机器抓出来的那个 intercept 截距项！

# 2. 拟合模型
final_fitted = final_model.fit(disp=False)

# 3. 往后精准预测未来 30 天
forecast_steps = 30
final_pred = final_fitted.forecast(steps=forecast_steps)

# 4. 绘制终极精准备货蓝图
plt.figure(figsize=(12, 5))
# 画出历史最后的 60 天销量作为对比
plt.plot(cleaned_df.index[-60:], cleaned_df['sales'].iloc[-60:], label='Historical Sales (Real)', color='blue')

# 挂上未来 30 天的红色预测虚线
future_dates = pd.date_range(start=cleaned_df.index[-1] + pd.Timedelta(days=1), periods=forecast_steps)
plt.plot(future_dates, final_pred, label='SARIMA (3,1,2)x(2,0,2)[7] Forecast', color='red', linestyle='--', linewidth=2.5)

plt.title("Ultimate Store Sales Demand Forecasting Blueprint")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import lightgbm as lgb

print("🚀 启动高级多表联动与时序滞后特征工程...")

# 1. 加载多源异构数据
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
oil = pd.read_csv('oil.csv')
stores = pd.read_csv('stores.csv')
holidays = pd.read_csv('holidays_events.csv')

# 统一合并处理
df_all = pd.concat([train, test], axis=0, ignore_index=True)
df_all['date'] = pd.to_datetime(df_all['date'])

# 2. 宏观经济指标：油价的前向填充与双阶滚动
oil['date'] = pd.to_datetime(oil['date'])
oil = oil.set_index('date').reindex(pd.date_range(df_all['date'].min(), df_all['date'].max())).ffill()
oil = oil.reset_index().rename(columns={'index': 'date'})
oil['oil_roll_mean_7'] = oil['dcoilwtico'].rolling(window=7, min_periods=1).mean()
df_all = df_all.merge(oil[['date', 'oil_roll_mean_7']], on='date', how='left')

# 3. 门店与地理特征融合
df_all = df_all.merge(stores, on='store_nbr', how='left')

# 4. 精准节假日联动 (仅保留未转移的有效全国性节日作为宏观冲击特征)
holidays['date'] = pd.to_datetime(holidays['date'])
national_holidays = holidays[(holidays['transferred'] == False) & (holidays['locale'] == 'National')].drop_duplicates(subset=['date'])
df_all = df_all.merge(national_holidays[['date', 'type']], on='date', how='left').rename(columns={'type': 'holiday_type'})
df_all['is_holiday'] = df_all['type'].notna().astype(int)

# 5. 基础时序时间戳拆解
df_all['dayofweek'] = df_all['date'].dt.dayofweek
df_all['month'] = df_all['date'].dt.month
df_all['is_weekend'] = df_all['dayofweek'].isin([5, 6]).astype(int)

# 6. 【高阶优化：构建时序滞后特征（Lag Features）】
# 因为测试集要预测未来 16 天，为了防止测试集没有 Lag 可用，滞后阶数必须从第 16 天开始往前推
print("正在构建时序记忆与滚动统计特征...")
df_all = df_all.sort_values(['store_nbr', 'family', 'date']).reset_index(drop=True)

# 创造历史销量的回音特征
for lag in [16, 17, 23]: # 16天前，17天前，以及两周前的同一天（23 = 16 + 7）
    df_all[f'sales_lag_{lag}'] = df_all.groupby(['store_nbr', 'family'])['sales'].shift(lag)

# 7. 类别特征编码转换
categorical_cols = ['family', 'city', 'state', 'holiday_type']
for col in categorical_cols:
    df_all[col] = df_all[col].astype('category')

# 8. 切分训练与测试，并进行【核心目标值对数变换】
train_eng = df_all[df_all['sales'].notna()].copy()
test_eng = df_all[df_all['sales'].isna()].copy()

features = [
    'store_nbr', 'family', 'onpromotion', 'cluster', 'oil_roll_mean_7', 
    'is_holiday', 'dayofweek', 'month', 'is_weekend', 'city', 'state', 'holiday_type',
    'sales_lag_16', 'sales_lag_17', 'sales_lag_23'
]

X_train = train_eng[features]
# 对目标值取对数，完美契合 RMSLE 指标，消除长尾偏态分布
y_train_log = np.log1p(train_eng['sales']) 
X_test = test_eng[features]

print("🏋️ 启动 LightGBM 迭代训练...")
gbm = lgb.LGBMRegressor(
    objective='regression',
    n_estimators=200,
    learning_rate=0.05,
    num_leaves=31,
    random_state=42,
    n_jobs=-1
)

gbm.fit(X_train, y_train_log, categorical_feature=categorical_cols)

# 9. 预测并进行【指数还原】
print("正在泛化预测并进行逆对数转换...")
preds_log = gbm.predict(X_test)
final_preds = np.expm1(preds_log) # 还原为真实销量
final_preds = np.clip(final_preds, 0, None)

# 10. 导出完美对齐官方 28512 行的最终文件
submission = pd.read_csv('sample_submission.csv')
submission['sales'] = final_preds
submission.to_csv('submission.csv', index=False)

print("\n🏆 【终极精雕细琢版】submission.csv 已完美导出！")
print(submission.head())
```

```python
import pandas as pd
import numpy as np
import lightgbm as lgb

print("🚀 启动高级多表联动与时序滞后特征工程...")

# 1. 重新从本地加载所有相关的原始多维数据表
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
oil = pd.read_csv('oil.csv')
stores = pd.read_csv('stores.csv')
holidays = pd.read_csv('holidays_events.csv')

# 统一合并训练集和测试集进行全局特征构建
df_all = pd.concat([train, test], axis=0, ignore_index=True)
df_all['date'] = pd.to_datetime(df_all['date'])

# 2. 宏观油价处理：补全非交易日空值，并计算 7 天移动平均
oil['date'] = pd.to_datetime(oil['date'])
oil = oil.set_index('date').reindex(pd.date_range(df_all['date'].min(), df_all['date'].max())).ffill()
oil = oil.reset_index().rename(columns={'index': 'date'})
oil['oil_roll_mean_7'] = oil['dcoilwtico'].rolling(window=7, min_periods=1).mean()
df_all = df_all.merge(oil[['date', 'oil_roll_mean_7']], on='date', how='left')

# 3. 融合门店区域特征
df_all = df_all.merge(stores, on='store_nbr', how='left')

# 4. 【核心修复】：基于官方格式，精准过滤并级联全国性有效节假日
holidays['date'] = pd.to_datetime(holidays['date'])
# 排除被转移的无效节日，并锁定 locale == 'National'
national_hols = holidays[(holidays['transferred'] == False) & (holidays['locale'] == 'National')].copy()
national_hols['is_national_holiday'] = 1
national_hols = national_hols.drop_duplicates(subset=['date'])

# 合并进大表
df_all = df_all.merge(national_hols[['date', 'is_national_holiday']], on='date', how='left')
df_all['is_national_holiday'] = df_all['is_national_holiday'].fillna(0).astype(int)

# 5. 时间戳细节拆解
df_all['dayofweek'] = df_all['date'].dt.dayofweek
df_all['month'] = df_all['date'].dt.month
df_all['is_weekend'] = df_all['dayofweek'].isin([5, 6]).astype(int)

# 6. 构建高阶时序滞后特征（Lag Features）
print("正在构建时序记忆与跨周滞后特征...")
df_all = df_all.sort_values(['store_nbr', 'family', 'date']).reset_index(drop=True)

# 创造历史销量的回音特征（严格从第 16 天往前推，完美满足未来 16 天预测期的可用性）
for lag in [16, 17, 23]:
    df_all[f'sales_lag_{lag}'] = df_all.groupby(['store_nbr', 'family'])['sales'].shift(lag)

# 7. 严格转换类别特征类型
categorical_cols = ['family', 'city', 'state']
for col in categorical_cols:
    df_all[col] = df_all[col].astype('category')

# 8. 切分训练与测试，并进行【目标值对数平滑变换】（完美对齐官方 RMSLE 评测标准）
train_eng = df_all[df_all['sales'].notna()].copy()
test_eng = df_all[df_all['sales'].isna()].copy()

# 确定核心特征池
features = [
    'store_nbr', 'family', 'onpromotion', 'cluster', 'oil_roll_mean_7', 
    'is_national_holiday', 'dayofweek', 'month', 'is_weekend', 'city', 'state',
    'sales_lag_16', 'sales_lag_17', 'sales_lag_23'
]

X_train = train_eng[features]
y_train_log = np.log1p(train_eng['sales'])  # 消解零售长尾偏态分布
X_test = test_eng[features]

print("🏋️ 正在进行大规模面板树模型拟合...")
gbm = lgb.LGBMRegressor(
    objective='regression',
    n_estimators=200,
    learning_rate=0.05,
    num_leaves=31,
    random_state=42,
    n_jobs=-1
)

gbm.fit(X_train, y_train_log, categorical_feature=categorical_cols)

# 9. 泛化预测并进行【指数还原】
print("正在执行预测并还原至真实销量空间...")
preds_log = gbm.predict(X_test)
final_preds = np.expm1(preds_log)
final_preds = np.clip(final_preds, 0, None)  # 依据商业逻辑约束负数销量

# 10. 读入官方模板，100% 对齐 28512 行导出交付件
submission = pd.read_csv('sample_submission.csv')
submission['sales'] = final_preds
submission.to_csv('submission.csv', index=False)

print("\n🏆 【完美全特征版】submission.csv 已大功告成！")
print(f"确认导出总行数: {len(submission)} (严丝合缝对齐 28512 行！)")
print(submission.head())
```
